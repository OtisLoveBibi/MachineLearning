```python
xTrain = trainSet.drop('median_house_value', axis=1)
yTrain = trainSet['median_house_value'].copy()
```

```python
xTrainNum = xTrain.drop('ocean_proximity', axis=1)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
xTrainNumImp = imputer.fit_transform(xTrainNum)
imputer.statistics_
```

```python
xTrainCat = xTrain[['ocean_proximity']]
xTrainCat.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
xTrainCatOhe = ohe.fit_transform(xTrainCat)
xTrainCatOhe.toarray()
```

```python
ohe.categories_
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

pipNum = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('stdScaler', StandardScaler())
])

pipFull = ColumnTransformer([
    ('num', pipNum, list(xTrain)[:8]),
    ('cat', OneHotEncoder(), [list(xTrain)[-1]])
])

xTrainPrep = pipFull.fit_transform(xTrain)
xTrainPrep.shape
```

```python
#
#LinearRegression

from sklearn.linear_model import LinearRegression

linReg = LinearRegression()
linReg.fit(xTrainPrep, yTrain)
linReg.score(xTrainPrep, yTrain)
```

```python
#DecisionTreeRegressor

from sklearn.tree import DecisionTreeRegressor

treeReg = DecisionTreeRegressor(random_state=1)
treeReg.fit(xTrainPrep, yTrain)
treeReg.score(xTrainPrep, yTrain)
```

```python
#RandomForestRegressor

from sklearn.ensemble import RandomForestRegressor

rfReg = RandomForestRegressor(random_state=1)
rfReg.fit(xTrainPrep, yTrain)
rfReg.score(xTrainPrep, yTrain)
```

```python
#

from sklearn.model_selection import cross_val_score
```

```python
#LinearRegression

linReg = LinearRegression()
linScores = cross_val_score(linReg, xTrainPrep, yTrain, scoring='r2', cv=10)
linScores
```

```python
linScores.mean(), linScores.std()
```

```python
#DecisionTreeRegressor

treeReg = DecisionTreeRegressor(random_state=1)
treeScores = cross_val_score(treeReg, xTrainPrep, yTrain, scoring='r2', cv=10)
treeScores
```

```python
treeScores.mean(), treeScores.std()
```

```python
#RandomForestRegressor

rfReg = RandomForestRegressor(random_state=1)
rfScores = cross_val_score(rfReg, xTrainPrep, yTrain, scoring='r2', cv=10)
rfScores
```

```python
rfScores.mean(), rfScores.std()
```

```python
#

#Random Search

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
paramDistribs = {
    'n_estimators': randint(low=1, high=200),
    'max_features': randint(low=1, high=8)
}

rndReg = RandomForestRegressor(random_state=1)
rndSearch = RandomizedSearchCV(rndReg, param_distributions=paramDistribs, n_iter=10, cv=5, scoring='r2', random_state=1)

rndSearch.fit(xTrainPrep, yTrain)
rndSearch.best_params_
```

```python
rlt = rndSearch.cv_results_
for mean_score, params in zip(rlt['mean_test_score'], rlt['params']):
    print(mean_score, params)
```

```python
#

bestModel = rndSearch.best_estimator_
bestModel
```

```python
#

#

xTest = testSet.drop('median_house_value', axis=1)
yTest = testSet['median_house_value'].copy()
```

```python
xTestPrep = pipFull.transform(xTest)
bestModel.score(xTestPrep, yTest)
```

```python
#

imp = bestModel.feature_importances_
imp
```

```python
catOneHotAttribs = list(pipFull.named_transformers_['cat'].categories_[0])
features = list(xTrain)[:8] + catOneHotAttribs
sorted(zip(imp, features), reverse=True)
```

```python
ind = np.argsort(imp)
plt.title('Feature Importances')
plt.barh(range(len(ind)), imp[ind], color='g', align='center')
plt.yticks(range(len(ind)), [features[i] for i in ind])
plt.xlabel('Relative Importances')
plt.show()
```

```python
#

#

pipFinal = Pipeline([
    ('preparation', pipFull),
    ('linear', bestModel)
])
pipFinal.fit(xTrain, yTrain)

finalModel = pipFinal
```

```python
#

import joblib
joblib.dump(finalModel, 'finalModelReg.pkl')
```
