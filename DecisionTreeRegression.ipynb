```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
m = 200
x = np.random.rand(m, 1)
y = 4 * (x- .5)**2 + np.random.randn(m, 1)/10
```

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y, 'b.')
plt.xlabel('$x_1$', fontsize=18)
plt.ylabel('y', fontsize=18, rotation=0)
plt.show()
```

```python
#DecisionTreeRegressor()

from sklearn.tree import DecisionTreeRegressor

treeReg = DecisionTreeRegressor(max_depth=2, random_state=1)
treeReg.fit(x, y)
```

```python
treeReg.score(x, y)
```

```python
#Tree plot

from sklearn import tree
plt.figure(figsize=(5, 3))
tree.plot_tree(treeReg, feature_names=['x1'], rounded=True, filled=True)
plt.savefig('treeReg.pdf')
plt.show()
```

```python
#Predict/Feature importances

treeReg.predict([[0.4]])
```

```python
treeReg.feature_importances_
```

```python
#Model

def plotRegressionPredictions(treeReg, x, y, axes=[0, 1, -0.2, 1]):
    x1 = np.linspace(axes[0], axes[1], 500).reshape(-1, 1)
    yPred = treeReg.predict(x1)
    
    plt.plot(x, y, 'b.')
    plt.plot(x1, yPred, 'r-', linewidth=2, label='$\hat{y}$')
    
    plt.axis(axes)
    plt.xlabel('$x_1$', fontsize=18)
    plt.ylabel('y', fontsize=18, rotation=0)
    plt.legend(loc='upper center', fontsize=14)
```

```python
plt.figure(figsize=(5, 3))
plotRegressionPredictions(treeReg, x, y)
plt.show()
```
