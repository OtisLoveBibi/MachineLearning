```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1) #一個固定的隨機參數

m = 100 #設定100個樣本數

x1 = 50 + 30 * np.random.rand(m,1) #從0到1隨機生成100 * 1的array
y = 135 + 0.5 * x1 + 3 * np.random.randn(m,1) #從0到1常態分佈生成100 * 1的array
```

```python
x1[:5]   #從x1取出前五項的數據
```

```python
y[:5]    #從y取出前五項的數據
```

```python
plt.figure(figsize=(5,4))    #設定圖型大小

plt.plot(x1, y ,"b.")   #在圖上標示出藍點

plt.xlabel("$x_1$",fontsize=18)    #設定Ｘ軸座標名稱與字體大小
plt.ylabel("$y$", rotation=0, fontsize=18)    #設定Ｙ軸座標名稱與大小及字體方向

plt.savefig("plot_ex1.pdf", dpi=300, bbox_inches='tight')     #將圖片存檔並設定解析度

plt.show()
```

```python
from sklearn.linear_model import LinearRegression    #載入函數

lin_reg = LinearRegression()   #將函數自定義命名
lin_reg.fit(x1, y)      #將x1跟y的數據使用函數運算
```

```python
lin_reg.intercept_, lin_reg.coef_    #顯示b與w1
```

```python
x1_new = np.array([[50],[80]])      #設定兩筆新的資料

lin_reg.predict(x1_new)      #用上述的模型預測新的資料的結果並得出y
```

```python
plt.figure(figsize=(5,4))

x1s = np.linspace(x1.min(),x1.max(),10).reshape(-1,1)       #設定迴歸線的x座標：從x1的最小值到最大值中間取10個數
y_pred = lin_reg.predict(x1s)     #將迴歸線的x座標放入模型計算出回歸線的y座標

plt.plot(x1, y, "b.")
plt.plot(x1s, y_pred, "r-", linewidth=2, label="$\hat y$")         #在圖上畫出回歸線並設定為紅線以及線地寬度與名稱

plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)   #設定回歸線的標示圖於圖片左上角並設定大小

plt.show()
```
